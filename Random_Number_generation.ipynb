Law Of Large Numbers 


Here I calculated mean and standard deviation for better understanding
and I learned how to store the random values in a array.

```python
import numpy as np
from numpy.random import randn
```

```python
count=0
N=1000
sd_count=0
all_values = np.empty((0, N))
mean_value = 0
for i in randn(N):
    all_values = np.append(all_values,[i])
    sd_count=sd_count+1;
    if i<1 and i>-1:
        #print('i is less than 1'+ str(i))
        count=count+1
        mean_value = round(mean_value +i,5)
        #print('The value of i '+ str(round(i,2)) + ' count '+str(count) +' mean_value ' + str(mean_value))
    #print(str(sd_count)+' I value '+str(round(i,2)) )
print('The Final Value = ' + str(count/N))
print('mean of random samples = ' + str(round(mean_value/N)))
print('printing all the values generated by random function \n'+str(all_values))
print('standard deviation = '+str(round(np.std(all_values),2)))
```
